```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset
df = pd.read_csv("taxis.csv")

# Vista r√°pida
df.head()
```

```python
# Informaci√≥n general
df.info()

# Nulos por columna
print(df.isnull().sum())

# Estad√≠sticas b√°sicas
df.describe()
```

```python
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
```

```python
df["trip_duration_min"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]).dt.total_seconds() / 60
```

```python
df.head()
```

```python
df = df[(df["trip_distance"] > 0) & 
        (df["total_amount"] > 0) & 
        (df["trip_duration_min"] > 0)]
```

```python
df.info()
```

```python
print("Tama√±o final del dataset:", df.shape)
print("Duraci√≥n media del viaje (min):", df["trip_duration_min"].mean())
print("Distancia media (millas):", df["trip_distance"].mean())
print("Importe medio ($):", df["total_amount"].mean())
```

```python
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].map({'N':0, 'Y':1})
```

```python
df.head(
)
```

```python
import matplotlib.pyplot as plt

# --- 1. Histograma de distancias ---
plt.figure(figsize=(8,5))
df["trip_distance"].plot.hist(bins=40, alpha=0.7)
plt.xlabel("Distancia (millas)")
plt.ylabel("N√∫mero de viajes")
plt.title("Distribuci√≥n de distancias de viaje (millas)")
plt.show()


# --- 2. Histograma de duraciones ---
plt.figure(figsize=(8,5))
df["trip_duration_min"].plot.hist(bins=40, alpha=0.7, color="orange")
plt.xlabel("Duraci√≥n (minutos)")
plt.ylabel("N√∫mero de viajes")
plt.title("Distribuci√≥n de duraciones de viaje (min)")
plt.show()


# --- 3. Relaci√≥n distancia vs importe total ---
plt.figure(figsize=(8,5))
plt.scatter(df["trip_distance"], df["total_amount"], alpha=0.5, s=10)
plt.xlabel("Distancia (millas)")
plt.ylabel("Importe total ($)")
plt.title("Distancia vs Importe Total")
plt.show()
```
